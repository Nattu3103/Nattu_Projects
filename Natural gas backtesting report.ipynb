```python
path="C:\\Users\\HP\\OneDrive\\Documents\\"
```

```python
df=pd.read_csv(path+"Natural_gas_backtested.csv")
#df.columns = df.iloc[0]
#df=df[1:]
df=df.iloc[:-2,:-3]
df
```

```python
df['Entry date']=pd.to_datetime(df['Entry date'])
df['entry_year']=df['Entry date'].dt.year
#df['year_month']=df['Entry date'].strftime("%Y-%M")
df
```

```python
PL_by_strategy=df.groupby(['entry_year','STRATEGY NAME'])[['PL_points']].sum()*1250
PL_by_strategy=PL_by_strategy.reset_index()
PL_by_strategy
```

```python

df=df.rename(columns={'P/L POINTS':'PL_points'})
df
```
