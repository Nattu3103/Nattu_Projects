```python
import pandas as pd
from datetime import date 

path='C:\\Users\\Siddhant\\Documents\\' # --> Specifying the path

df_amibroker_positions=pd.read_csv(path+'D7 Amibroker Positions.csv') # --> read_csv_file from amibroker software

```

```python
#dropping sepcific clients

df_amibroker_positions=df_amibroker_positions.drop(df_amibroker_positions[(df_amibroker_positions['Client']==11866) | (df_amibroker_positions['Client']==90197) |(df_amibroker_positions['Client']==9771) |(df_amibroker_positions['Client']==90245) |(df_amibroker_positions['Client']==13598)].index)

# changing datdtype of 'expiry date'

df_amibroker_positions['Expiry']=pd.to_datetime(df_amibroker_positions['Expiry'])

# arranging it in required format

df_amibroker_positions['Expiry']=df_amibroker_positions['Expiry'].dt.strftime('%d-%b-%Y')

```

```python
# confirming datatype change

print(df_amibroker_positions['Expiry'].info())

#adding new column 
df_amibroker_positions['Type']='AMI'

#printing
df_amibroker_positions
```

```python
# read positions from trading terminal

df1_ctcl_positions=pd.read_csv(path+'Net Position.txt',sep="\t")

df1_ctcl_positions
```

```python
# exporting it to csv

df1_ctcl_positions.to_csv(path+'Net_Position.csv')
df1_ctcl_positions
```

```python
# deleting rows with net qty 0
df1_ctcl_positions=df1_ctcl_positions[df1_ctcl_positions['NetQty']!=0]
```

```python
#print(df1_ctcl_positions)
df1_ctcl_positions.to_csv(path+'Net_Position.csv')
df1_ctcl_positions
```

```python

# Filtering out only the required columns
df1_ctcl_positions=df1_ctcl_positions[['Client','Ser_Exp','StrikePrice','OptionType','NetQty','Strategy','Symbol']]
df1_ctcl_positions
```

```python
# Renaming trading terminal columns as per amibroker software

df1_ctcl_positions=df1_ctcl_positions.rename(columns={'Ser_Exp':'Expiry','OptionType':'Option Type','StrikePrice':'Strike Price','NetQty':'Net Qty'})
df1_ctcl_positions
```

```python
# changing data type of Expiry column

df1_ctcl_positions['Expiry']=pd.to_datetime(df1_ctcl_positions['Expiry'])
df1_ctcl_positions['Expiry']=df1_ctcl_positions['Expiry'].dt.strftime('%d-%b-%Y')
```

```python
# adding a new column to identify the file is from trading terminal

df1_ctcl_positions['Type']='CTCL'
df1_ctcl_positions
```

```python
# Drop Nan values
df1_ctcl_positions=df1_ctcl_positions.dropna()

df1_ctcl_positions
```

```python
# concatenate two files and getting the resultant file

final_df=pd.concat([df_amibroker_positions,df1_ctcl_positions])
final_df
```

```python
# Dropping the duplicates to identify the difference in two csv files

output=final_df.drop_duplicates(subset=['Client','Expiry','Strike Price','Option Type','Net Qty','Strategy','Symbol'],keep=False)
output
```

```python
#Identifying and reporting if there is a difference

if len(output)>0:
    output.to_csv(path+'difference_file.csv')
    print('There is some difference')
    print(output['Strategy'].unique()) # unique to print distinct strategy that shows error
else:
    print('All positions are fine')
```
